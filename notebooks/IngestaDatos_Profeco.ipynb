# Extracción de datos de la PROFECO para los precios del Aguacte Hass

```python
import requests
import pandas as pd

# 1. Definimos la URL exacta que interceptaste
url = "https://qqp.profeco.gob.mx/api/precios?clave_ciudad=&busqueda=aguacate"

# 2. Replicamos los headers esenciales
headers = {
    "accept": "*/*",
    "accept-language": "es-MX,es-ES;q=0.9,es;q=0.8",
    "authorization": "Basic cXFwd2ViOnFxcHdlYg==",
    # Es buena práctica poner un User-Agent para simular ser un navegador
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# 3. Hacemos la petición GET
print("Realizando la petición a la API de PROFECO...")
response = requests.get(url, headers=headers)

# 4. Verificamos que la petición fue exitosa (Código 200)
if response.status_code == 200:
    # Convertimos la respuesta a JSON
    data = response.json()

    # IMPORTANTE: Aquí dependemos de la estructura del JSON.
    # Usualmente los datos vienen en una llave como 'data', 'resultados', o directamente es una lista.
    # Si 'data' es una lista de diccionarios, Pandas lo convierte directo:

    # df_profeco = pd.DataFrame(data['resultados']) # Ajusta 'resultados' según lo que imprima el JSON

    # Por ahora, vamos a imprimir las llaves del JSON para ver cómo viene estructurado
    if isinstance(data, dict):
        print("Estructura del JSON (llaves):", data.keys())
    else:
        print("El JSON es una lista con", len(data), "registros.")
        df_profeco = pd.DataFrame(data)
        display(df_profeco.head())

else:
    print(f"Error al conectar: {response.status_code}")
    print(response.text)
```
