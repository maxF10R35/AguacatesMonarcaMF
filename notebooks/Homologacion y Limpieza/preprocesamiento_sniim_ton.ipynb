```python
import pandas as pd
```

```python
df_ton = pd.read_excel('/content/datos_volumen_ton_sniim_2001_2026.xlsx')

df_ton.head()
```

```python
df_ton.info()
```

```python
print(f'Se van a analizar {df_ton.shape[0]} registros')
print('---' * 10)

#duplicados
print('Analizamos los datos duplicados')
print('---' * 10)
total_duplicados = df_ton.duplicated().sum()
print(f"Total de filas duplicadas: {total_duplicados}")

df_sin_dupes = df_ton.drop_duplicates()
print(f"Total de filas sin duplicados: {df_sin_dupes.shape[0]}")

porcentaje = (total_duplicados / df_ton.shape[0]) * 100
print(f"Porcentaje de filas duplicadas: {porcentaje:.2f}%")

pct_retencion = 100 - porcentaje
print(f"Porcentaje de filas retenidas: {pct_retencion:.2f}%")
```

```python
#Transformamos a tipo DateTime

df_sin_dupes['Fecha'] = pd.to_datetime(df_ton['Fecha'])

df_sin_dupes.info()
```

```python
from numpy import transpose
#Verificamos los tipos de presentaciones
presentaciones = df_sin_dupes['Presentacion'].unique()
print('--- Verificando tipos de presentaciones ---')
print(f'Hay {len(presentaciones)} presentaciones, las cuales son: {presentaciones}')

#verificamos productores
productores = df_sin_dupes['Origen'].unique()
print('--- Verificando productores ---')
print(f'Hay {len(productores)} estados que participan, que son: {productores}')

#Verificamos los tipos de transporte
transportes = df_sin_dupes['Transporte'].unique()
print('--- Verificando tipos de transporte ---')
print(f'Hay {len(transportes)} tipos de transporte, que son: {transportes}')


#Verificamos cuáles años tenemos
años = df_sin_dupes['Fecha'].dt.year.unique()
print('--- Verificando años ---')
print(f'Hay {len(años)} años, que son: {años}')
```

```python
#Revisamos datos nulos

df_sin_dupes.isna().sum()
```

```python
#guardamos

df_sin_dupes.to_csv('volumen_ton_sniim.csv', index=False, encoding='utf-8')
```
