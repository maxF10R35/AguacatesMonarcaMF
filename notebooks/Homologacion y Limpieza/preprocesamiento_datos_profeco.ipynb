```python
import pandas as pd
```

```python
df_sucio = pd.read_csv('/content/precios_competencia_profeco.csv')

df_sucio.head()
```

```python
df_sucio.tail()
```

```python
print(f'Se van a analizar {df_sucio.shape[0]} registros')
print('---' * 10)

#duplicados
print('Analizamos los datos duplicados')
print('---' * 10)
total_duplicados = df_sucio.duplicated().sum()
print(f"Total de filas duplicadas: {total_duplicados}")

df_sin_dupes = df_sucio.drop_duplicates()
print(f"Total de filas sin duplicados: {df_sin_dupes.shape[0]}")

porcentaje = (total_duplicados / df_sucio.shape[0]) * 100
print(f"Porcentaje de filas duplicadas: {porcentaje:.2f}%")

pct_retencion = 100 - porcentaje
print(f"Porcentaje de filas retenidas: {pct_retencion:.2f}%")
```

```python
print(f'Los tipos de datos son: \n{df_sin_dupes.dtypes}')
```

```python
#Cambiamos el formato a Date

df_sin_dupes['fecha_observacion'] = pd.to_datetime(df_sin_dupes['fecha_observacion'], errors='coerce')

print(f'Los tipos de datos son: \n{df_sin_dupes.dtypes}')

df_sin_dupes.head()
```

```python
df_sin_dupes.info()
```

```python
#Verificamos si solo tenemos aguacates
no_aguacate = df_sin_dupes[df_sin_dupes['tipo_producto'] != 'AGUACATE']
print('--- Verificando tipos de productos ---')
print(f'Hay {len(no_aguacate)} datos que no son de aguacate')

#verificamos las variedades y presentación de los aguacates
tipos = df_sin_dupes['producto'].unique()
print('--- Verificando tipos y presentaciones de los aguacates ---')
print(f'Hay {len(tipos)} variedades o presentaciones, que son: {tipos}')


#Verificamos los años que hay registrados
anios = df_sin_dupes['fecha_observacion'].dt.year.unique()
print('--- Verificando años ---')
print(f'Hay {len(anios)} años registrados: {anios}')

#observamos las entidades registradas
entidades = df_sin_dupes['entidad'].unique()
print('--- Verificando entidades ---')
print(f'Hay {len(entidades)} entidades registradas: {entidades}')

#Para la cdmx vemos los municipios
cdmx = df_sin_dupes[df_sin_dupes['entidad'] == 'Ciudad de México']
municipios = cdmx['municipio'].unique()
print('--- Verificando municipios ---')
print(f'Hay {len(municipios)} municipios registrados: {municipios}')
```

```python
#verificamos nulos por última vez

df_sin_dupes.isna().sum()
```

```python
#Reporte final
totales = df_sucio.shape[0]
final = df_sin_dupes.shape[0]
pct_ret = final / totales * 100
print(f'De {totales} registros, quedaron solo {final}')
print(f'El porcentaje de retención es del {round(pct_retencion, 2)}%')
```

```python
df_limpio = df_sin_dupes.sort_values(by='fecha_observacion')

df_limpio.head()
```

```python
#Limitamos a la cdmx

df_cdmx = df_limpio[df_limpio['entidad'] == 'Ciudad de México']

df_cdmx.head()
```

```python
print(f'Para la cdmx se encontraron {df_cdmx.shape[0]} datos')
```

```python
#guardamos el df limpio

df_cdmx.to_csv('competencia_cdmx_limpio.csv', index=False, encoding='UTF-8')
```
